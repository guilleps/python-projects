```python
def predecir_medalla(peso, edad):
    if edad > 24:
        if peso > 70:
            return "Sí"  # Medallista
        else:
            return "No"  # No medallista
    else:
        if peso < 65:
            return "Sí"  # Medallista
        else:
            return "No"  # No medallista

# Datos originales para comprobar el funcionamiento
datos = [
    {"peso": 90, "edad": 18},  # No
    {"peso": 80, "edad": 20},  # No
    {"peso": 70, "edad": 19},  # No
    {"peso": 70, "edad": 25},  # No
    {"peso": 60, "edad": 18},  # Sí
    {"peso": 80, "edad": 28},  # Sí
    {"peso": 85, "edad": 26},  # Sí
    {"peso": 90, "edad": 25},  # Sí
]

# Predicciones
for i, persona in enumerate(datos, start=1):
    prediccion = predecir_medalla(persona["peso"], persona["edad"])
    print(f"Persona {i}: Peso = {persona['peso']}, Edad = {persona['edad']} → Predicción: {prediccion}")

```
