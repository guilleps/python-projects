```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn import tree

# Datos proporcionados
data = {
    'Peso': [90, 80, 70, 70, 60, 80, 85, 90],
    'Edad': [18, 20, 19, 25, 18, 28, 26, 25],
    'Medalla': ['No', 'No', 'No', 'No', 'Sí', 'Sí', 'Sí', 'Sí']
}

# Crear DataFrame
df = pd.DataFrame(data)

# Convertimos la etiqueta 'Medalla' a binaria
df['Medalla'] = df['Medalla'].map({'No': 0, 'Sí': 1})

# Variables independientes y etiqueta
X = df[['Peso', 'Edad']]
y = df['Medalla']

# Crear el clasificador con profundidad máxima 2 para simular las divisiones que mencionaste
clf = DecisionTreeClassifier(max_depth=2, random_state=42)
clf.fit(X, y)

# Mostrar el árbol
plt.figure(figsize=(12, 8))
plot_tree(clf, feature_names=['Peso', 'Edad'], class_names=['No', 'Sí'], filled=True, rounded=True)
plt.title("Árbol de Decisión basado en Edad y Peso")
plt.show()

# Opcional: mostrar reglas en texto
rules = tree.export_text(clf, feature_names=['Peso', 'Edad'])
print("Reglas del árbol:\n")
print(rules)
```
